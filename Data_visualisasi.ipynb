```python
import pandas as pd

# Baca data
data = pd.read_csv('dataset3.csv')

# Tampilkan beberapa data awal untuk memastikan data terbaca dengan benar
data.head()
```

```python
import matplotlib.pyplot as plt

# Mengelompokkan data berdasarkan 'Product line' dan menghitung total 'cogs'
product_line_cogs = data.groupby('Product line')['cogs'].sum()

# Membuat diagram batang
plt.figure(figsize=(10, 6))
product_line_cogs.plot(kind='bar', color='skyblue')
plt.title("Total COGS per Product Line")
plt.xlabel("Product Line")
plt.ylabel("COGS")
plt.xticks(rotation=45)
plt.show()
```

```python
# Mengonversi kolom 'Date' menjadi format datetime untuk ekstraksi bulan
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')  # Specify the correct format
data['Month'] = data['Date'].dt.strftime('%Y-%m')  # Format tahun-bulan

# Mengelompokkan data berdasarkan 'Month' dan menghitung total 'cogs'
monthly_cogs = data.groupby('Month')['cogs'].sum()

# Membuat diagram garis
plt.figure(figsize=(10, 6))
monthly_cogs.plot(kind='line', marker='o', color='orange')
plt.title("Total COGS per Month")
plt.xlabel("Month")
plt.ylabel("Total COGS")
plt.xticks(rotation=45)
plt.grid(visible=True)
plt.show()
```

```python
# Menghitung jumlah metode pembayaran
payment_counts = data['Payment'].value_counts()

# Membuat diagram pie
plt.figure(figsize=(8, 8))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'salmon'])
plt.title("Distribution of Payment Methods")
plt.show()
```

```python
# Mengelompokkan data berdasarkan 'City' dan menghitung rata-rata 'Rating'
city_rating_avg = data.groupby('City')['Rating'].mean()

# Membuat diagram batang horizontal
plt.figure(figsize=(10, 6))
city_rating_avg.plot(kind='barh', color='purple')
plt.title("Average Rating by City")
plt.xlabel("Average Rating")
plt.ylabel("City")
plt.show()
```

```python
import seaborn as sns

# Membuat box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Product line', y='Total', data=data)
plt.title("Total Distribution by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.show()
```
