```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

doc1 = "I love machine learning"
doc2 = "I enjoy learning machine"

vectorizer = CountVectorizer()
X = vectorizer.fit_transform([doc1, doc2])

cos_sim = cosine_similarity(X[0:1], X[1:2])
print(X.toarray())
print(f"Cosine similarity :{cos_sim[0][0]}")

doc1_vec = X[0].toarray()[0]
doc2_vec = X[1].toarray()[0]

plt.figure(figsize=(8, 6))

plt.quiver(0, 0, doc1_vec[0], doc1_vec[1], angles='xy', scale_units='xy',scale=1, color='blue', label="Dokumen 1")
plt.quiver(0, 0, doc2_vec[0], doc2_vec[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Dokumen 2")

plt.text(doc1_vec[0] + 0.1, doc1_vec[1] + 0.1, "Dokumen 1", color="blue")
plt.text(doc2_vec[0] + 0.1, doc2_vec[1] + 0.1, "Dokumen 2", color="red")

plt.xlim(-1, max(doc1_vec[0], doc2_vec[0]) + 2)
plt.ylim(-1, max(doc1_vec[1], doc2_vec[1]) + 2)
plt.xlabel('Kata 1')
plt.ylabel('Kata 2')
plt.title(f"Cosine Similarity: {cos_sim[0][0]:.2f}")
plt.legend()
plt.grid(True)
plt.show()
```
