```python
rev = df['revenues']
print (rev)
```

```python
import re
rev_replace = rev.str.replace(r'\$', '', regex=True)
print(rev_replace)
```

```python
revenues = rev_replace.astype(float)
print(revenues)
```

```python
mean = revenues.mean()
print(mean)
```

```python
# prompt: make median

median = revenues.median()
median
```

```python
# prompt: make modus

modus = revenues.mode()
modus
```

```python
data = revenues
print(data)
```

```python
sns.histplot(data, kde=True)
```

```python
# prompt: get quartil 1,2,3, standar deviasi,min , max, variansi, inter quartil

# Calculate quartiles, standard deviation, min, max, variance, and interquartile range
q1 = revenues.quantile(0.25)
q2 = revenues.quantile(0.50)  # Median
q3 = revenues.quantile(0.75)
std_dev = revenues.std()
min_val = revenues.min()
max_val = revenues.max()
variance = revenues.var()
iqr = q3 - q1

print(f"Q1: {q1}")
print(f"Q2 (Median): {q2}")
print(f"Q3: {q3}")
print(f"Standard Deviation: {std_dev}")
print(f"Minimum: {min_val}")
print(f"Maximum: {max_val}")
print(f"Variance: {variance}")
print(f"Interquartile Range (IQR): {iqr}")
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
sns.boxplot(y=revenues)  # Assuming 'revenues' is your data
plt.title('Box Plot Pendapatan')
plt.ylabel('Harga')
plt.show()
```

```python
lower_bound = q2 - 1.5 * iqr

print(f"Lower Bound: {lower_bound}")
```

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

data_outliers = revenues[(revenues < lower_bound) | (revenues > upper_bound)]
```
